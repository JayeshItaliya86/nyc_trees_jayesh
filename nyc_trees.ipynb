```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Read the csv file of our data set
data_trees = pd.read_csv('./data_100000.csv', skipinitialspace=True)
df = data_trees.copy() # Make a copy of data set
df.head() 
```

```python
# To see the columns names
df.columns
```

```python
# Drop some columns
df = df.drop(columns=['the_geom', 'stump_diam', 'spc_latin',
                      'problems', 'address', 'zip_city', 'cb_num',
                      'borocode', 'boroname', 'cncldist',
                      'st_assem', 'st_senate', 'nta', 'boro_ct', 
                      'state', 'x_sp', 'y_sp'], axis=1)
```

```python
# create the dictionary with the old and new names
new_names = {'tree_dbh': 'tree_diameter', 'curb_loc': 'curb_location',
             'spc_common': 'species_common_name', 'trnk_wire': 'trunk_wire',
             'trnk_light': 'trunk_light', 'trnk_other': 'trunk_other',
             'brnch_ligh': 'branch_light', 'brnch_shoe': 'branch_sneakers',
             'brnch_othe': 'branch_other'}
print(new_names)
print(type(new_names))
```

```python
# Rename the columns
df = df.rename(columns=new_names)

print(df.columns)
```

```python
# To identify duplicate values in our data set
duplicate_rows = df[df.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicate_rows)
```

```python
# To see the statistics values of data set
df.describe()
```

```python
# To check how many values are per columns unique with the help of function
def value_count(data):
    for item in range(len(data)):
        print(f"Value counts from column ** {df.columns[item]} ** returns: \n {df[df.columns[item]].value_counts()}\n")

value_count(df.columns)
```

```python
df.dtypes
```

```python
# Change created_at data type from object to datetime
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df.dtypes
```

```python
# Fill a missing values 

df['health'] = df['health'].replace(" ", np.NaN)
df['species_common_name'] = df['species_common_name'].replace(" ", np.NaN)
df['steward'] = df['steward'].replace(" ", np.NaN)
df['guards'] = df['guards'].replace(" ", np.NaN)
df['sidewalk'] = df['sidewalk'].replace(" ", np.NaN)
```

```python
df.isnull().sum()
```

```python
# To change all categorical values into title case
df = df.applymap(lambda value: value.title() if type(value) == str else value)
```

```python
# setting to display all columns
pd.set_option("Display.max_columns", None)
```

```python
def change_classification(data):
    if data == 'Yes':
        return 1
    elif data == 'No':
        return 0
```

```python
df['root_stone'] = df['root_stone'].apply(lambda data: change_classification(data))
df['root_grate'] = df['root_grate'].apply(lambda data: change_classification(data))
df['root_other'] = df['root_other'].apply(lambda data: change_classification(data))
df['trunk_wire'] = df['trunk_wire'].apply(lambda data: change_classification(data))
df['trunk_light'] = df['trunk_light'].apply(lambda data: change_classification(data))
df['trunk_other'] = df['trunk_other'].apply(lambda data: change_classification(data))
df['branch_light'] = df['branch_light'].apply(lambda data: change_classification(data))
df['branch_sneakers'] = df['branch_sneakers'].apply(lambda data: change_classification(data))
df['branch_other'] = df['branch_other'].apply(lambda data: change_classification(data))
```

```python
# Round the decimal for latitude and longitude
columns_round = ['latitude', 'longitude']
df[columns_round] = df[columns_round].round(8)
```

```python
# Add leading zeros to the integer columns

df['tree_diameter'] = df['tree_diameter'].apply(lambda x: '{0:0>2}'.format(x))
```
